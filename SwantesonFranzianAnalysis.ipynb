```python
 
def severityStack(rasterLoc, runLength):

    rasterlist = []
    timesteps = np.linspace(1,runLength,runLength)

    repnames = next(os.walk(rasterLoc))[1]
    num_replicates = len(repnames)

    idx = 0
    for rep in repnames:
        #print rep
        counter = 0
        repnum = rep.split('replicate')[1]
        for time in timesteps:  
            wkDir_data = rasterLoc + '/' 'replicate' + repnum + '/' + 'firesevmap/'
            sevmap = 'severity-' + str(int(time)) + '.gis'
            src_ds = gdal.Open( wkDir_data + sevmap ) 
            sevarray = src_ds.ReadAsArray()
            sevarray = sevarray.astype('float')
            rasterlist.append(sevarray)
    return rasterlist

def severityGen(rasterList):
    rasterStack = np.dstack(rasterList)  
    rasterAdj = rasterStack - 2
    rasterAdj[rasterAdj < 0] = np.nan
    meanSev = np.nanmean(rasterAdj, axis = 2)
    varSev = np.nanvar(rasterAdj, axis = 2)
    return meanSev, varSev
```

```python
def plotAllSimsMinMaxMed_2(NECNdf, var, ax = None, style = None, 
                           fillcolor=None, label = None, *args, **kwargs):  
    # Create the min, max, and median DFs
    summed = NECNdf.groupby(['Time','rep','Sim']).sum()
    fillcol = fillcolor
    # Reset the index for grouping
    reset_summed = summed.reset_index()
    
    maxdf = reset_summed.groupby('Time').max()
    mindf = reset_summed.groupby('Time').min()
    meandf = reset_summed.groupby('Time').mean()

    
    index = pd.date_range('2000-1-1', periods=len(maxdf), freq='1A')
    for df in [maxdf, mindf, meandf]:
        df.index = index

    # Add the three plots to a single set of axes
    # Grab figure axes if none stated
    if ax == None:
         ax = plt.gca()
    
    # Plot lines for min, max, and median
    #maxdf[var].plot(ax = ax, lw = 1.5, color = None, label = '_nolegend_', *args, **kwargs)
    #mindf[var].plot(ax = ax, lw = 1.5, color = None, label = '_nolegend_',  *args, **kwargs)
    rolmean = meandf.rolling(3, win_type='triang').mean()
    rolmean[var].plot(ax = ax, lw = 1.5, ls = style, label = label, color = 'black', *args, **kwargs)
    
    # Fill between min and max
    ax.fill_between(maxdf.index, maxdf[var], mindf[var], color = fillcol, alpha = 0.2)
```

```python
def plotAllSimsMinMaxMed(NECNdf, var, ax = None, fillcolor=None, *args, **kwargs):  
    # Create the min, max, and median DFs
    maxdf = NECNdf.groupby('Time').max()
    mindf = NECNdf.groupby('Time').min()
    meandf = NECNdf.groupby('Time').mean()
    fillC = fillcolor
    
    index = pd.date_range('2000-1-1', periods=len(maxdf), freq='1A')
    for df in [maxdf, mindf, meandf]:
        df.index = index

    # Add the three plots to a single set of axes
    # Grab figure axes if none stated
    if ax == None:
         ax = plt.gca()
    
    # Plot lines for min, max, and median
    maxdf[var].plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    mindf[var].plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    rolmean = meandf.rolling(3, win_type='triang').mean()
    rolmean[var].plot(ax = ax, lw = 1.5, ls = '-', color = 'black', *args, **kwargs)
    
    # Fill between min and max
    ax.fill_between(maxdf.index, maxdf[var], mindf[var], color = fillC, alpha = 0.2)
    
def plotAllSimsMinMaxMedSUM(NECNdf, var, ax = None, *args, **kwargs):  
    # Create the min, max, and median DFs
    allreps = NECNdf.groupby(['Time', 'Sim', 'rep']).mean().reset_index()
    mindf = NECNdf.groupby(['Time']).min()
    maxdf = NECNdf.groupby(['Time']).max()
    meandf = NECNdf.groupby(['Time']).mean()
  
    maxdfS = maxdf[var].cumsum()
    mindfS = mindf[var].cumsum()
    meandfS = meandf[var].cumsum()
   
    index = pd.date_range('2000-1-1', periods=len(maxdfS), freq='1A')
    for df in [maxdfS, mindfS, meandfS]:
        df.index = index

    # Add the three plots to a single set of axes
    # Grab figure axes if none stated
    if ax == None:
         ax = plt.gca()
    
    # Plot lines for min, max, and median
    maxdfS.plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    mindfS.plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    meandfS.plot(ax = ax, lw = 1.5, color = 'black', ls = '-', *args, **kwargs)
    
    # Fill between min and max
    ax.fill_between(maxdfS.index, maxdfS, mindfS, color = 'blue', alpha = 0.2)
    
    
def plotAllSimsMinMaxMedSUM2(NECNdf, var, ax = None, *args, **kwargs):  
    # Create the min, max, and median DFs
    maxdf = NECNdf.groupby('Time').max()
    mindf = NECNdf.groupby('Time').min()
    meandf = NECNdf.groupby('Time').mean()

    maxdfS = maxdf[var].cumsum()
    mindfS = mindf[var].cumsum()
    meandfS = meandf[var].cumsum()
   
    index = pd.date_range('2000-1-1', periods=len(maxdfS), freq='1A')
    for df in [maxdfS, mindfS, meandfS]:
        df.index = index

    # Add the three plots to a single set of axes
    # Grab figure axes if none stated
    if ax == None:
         ax = plt.gca()
    
    # Plot lines for min, max, and median
    maxdfS.plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    mindfS.plot(ax = ax, lw = 1.5, color = 'white', *args, **kwargs)
    meandfS.plot(ax = ax, lw = 1.5, color = 'black', ls = '--', *args, **kwargs)
    
    # Fill between min and max
    ax.fill_between(maxdfS.index, maxdfS, mindfS, color = 'green', alpha = 0.2)

```
